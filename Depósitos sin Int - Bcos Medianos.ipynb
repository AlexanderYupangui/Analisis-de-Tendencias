#### Importar Librer铆as

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm
from scipy.stats import normaltest
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score

sns.set_theme(style = "darkgrid")

pd.set_option('display.float_format', '{:.2f}'.format)
```

#### Carga de Datos

```python
df = pd.read_excel("Indicadores.xlsx").set_index('FECHA')
df.head()
```

##  Preprocesamiento de Datos

```python
## Agrupaci贸n de Datos por Mes y por Tipo
df = df.drop(columns = ["CODIGO", "CUENTA"])
df = df.groupby(["FECHA", "NOMBRE_CORTO"]).sum()

## Agrupaci贸n de Bancos por su tama帽o en {Grandes, Medianos y Chicos}
bp_grandes = ["BP_GUAYAQUIL", "BP_PACIFICO", "BP_PICHINCHA", "BP_PRODUBANCO"]
bp_medianos = ["BP_AUSTRO", "BP_BOLIVARIANO", "BP_CITIBANK", "BP_DINERS", "BP_GENERAL_RUMIAHUI", "BP_INTERNACIONAL", "BP_LOJA", "BP_MACHALA", "BP_SOLIDARIO"]
bp_chicos = ["BP_PROCREDIT", "BP_AMAZONAS", "BP_COMERCIAL_DE_MANABI", "BP_LITORAL", "BP_COOPNACIONAL", "BP_CAPITAL", "BP_FINCA", "BP_DELBANK", "BP_DMIRO_SA", "BP_BANCODESARROLLO", "BP_VISIONFUND_ECUADOR"]

df["BP_GRANDES"] = df[bp_grandes].sum(axis = 1)
df["BP_MEDIANOS"] = df[bp_medianos].sum(axis = 1)
df["BP_CHICOS"] = df[bp_chicos].sum(axis = 1)

df = df.drop(columns = bp_grandes)
df = df.drop(columns = bp_medianos)
df = df.drop(columns = bp_chicos)

df = df.reset_index()

## Proceso de descomposici贸n y pivoteo de dataset
df_procesado = df.melt(id_vars = ["FECHA", "NOMBRE_CORTO"], value_vars = ["BP_GRANDES", "BP_MEDIANOS", "BP_CHICOS"], var_name = "CUENTA", value_name = "MONTO")
df_procesado["TIPO_CUENTA"] = df_procesado["NOMBRE_CORTO"] + '_' + df_procesado["CUENTA"]

df = df_procesado.pivot(index = "FECHA", columns = "TIPO_CUENTA", values = "MONTO").reset_index()

df.index.name = 'TIPO_CUENTA'
df = df.reset_index()
df.columns.name = None
df = df.drop(columns = ["TIPO_CUENTA"])

df.head()
```

#### An谩lisis de datos

```python
variable = "Dep贸sito sin Inter茅s_BP_MEDIANOS"
variables = ["FECHA", "Dep贸sito sin Inter茅s_BP_MEDIANOS"]

data = df[variables].dropna()

data["y"] = df[variable]
data = data.drop(columns = [variable])
data.head(5)
```

```python
# Lienzo para visualizar la Serie de Tiempo
plt.figure(figsize = (15, 4), dpi = 100)
sns.lineplot(x = "FECHA", y = "y", data = data)
plt.xlabel("Fecha")
plt.ylabel("Cartera")
plt.title("Dep贸sitos s/Int de Bancos Medianos")

plt.grid(True)
plt.tight_layout()
plt.show()
```

Estos valores pueden ser producto de errores de medici贸n o eventos no repetibles que no aportan informaci贸n relevante para el an谩lisis continuo. Por lo tanto, eliminarlos permite una representaci贸n m谩s precisa y coherente de los datos, mejorando la calidad del modelo y la fiabilidad de las predicciones.

## 1锔 Metodolog铆a Box-Jenkins

An谩lisis exploratorio:
1) La primera gr谩fica muestra la serie original.
2) La segunda gr谩fica muestra la tendencia de la serie. Al ser una curva creciente, se muestra que la serie tiene tendencia positiva (SERIE NO ESTACIONARIA)
3) La tercera gr谩fica muestra la estacionalidad de la serie, la cual S muestra un patr贸n estacional
4) La cuarta gr谩fica muestra los residuos de la serie, la cual muestra un patr贸n

```python
## Obtener Datos
data_arima = data.copy()
```

Prueba de estacionariedad mediante el test de Dickey-Fuller

```python
test_DF = adfuller(data_arima["y"], autolag = "AIC")
print(f"Estad铆stico ADF: {test_DF[0]}")
print(f"p-value: {test_DF[1]}")
if test_DF[1] <= 0.05:
    print("La serie S es estacionaria")
else:
    print("La serie NO es estacionaria")
```

Dado que la serie S es estacionaria en media, vamos a continuar con la Metodolog铆a

Aunque la prueba de estacionariedad de Dickey-Fuller indica que la serie es estacionaria, el an谩lisis de la Funci贸n de Autocorrelaci贸n (ACF) sugiere lo contrario, ya que la autocorrelaci贸n se reduce lentamente en lugar de truncarse r谩pidamente. Por esta raz贸n, se aplicar谩 una diferenciaci贸n adicional para garantizar la estacionariedad del modelo.

```python
# Aseg煤rate de que no haya valores NaN en la columna antes de graficar
serie = data_arima["y"]

# Crear subplots para ACF y PACF
fig, axes = plt.subplots(1, 2, figsize = (16, 6))

# Gr谩fico de la ACF
plot_acf(serie, lags = 20, ax = axes[0])
axes[0].set_title("Funci贸n de Autocorrelaci贸n para los Dep贸sitos s/Int de Bancos Medianos")

# Gr谩fico de la PACF
plot_pacf(serie, lags = 20, ax = axes[1])
axes[1].set_title("Funci贸n de Autocorrelaci贸n Parcial para los Dep贸sitos s/Int de Bancos Medianos")

# Mostrar las gr谩ficas
plt.tight_layout()
plt.show()
```

```python
data_arima["diff_y"] = data_arima["y"].diff().dropna()
data_arima = data_arima.dropna()
data_arima.head()
```

```python
# Lienzo para visualizar la Serie de Tiempo
plt.figure(figsize = (15, 4), dpi = 100)
sns.lineplot(x = "FECHA", y = "diff_y", data = data_arima)
plt.xlabel("Fecha")
plt.ylabel("Cartera")
plt.title("Primera Diferencia de Dep贸sitos s/Int de Bancos Medianos")

plt.grid(True)
plt.tight_layout()
plt.show()
```

La serie ya muestra buen comportamiento. Por lo cual, se continua con la estimaci贸n de los par谩metros del Modelo $$ARIMA (p, d, q)$$

#### An谩lisis de Funciones de Autocorrelaci贸n y Autocorrelaci贸n Parcial

El an谩lisis de la Funci贸n de Autocorrelaci贸n (ACF) y la Funci贸n de Autocorrelaci贸n Parcial (PACF) revela que ambas funciones se truncan en el rezago 1, lo que indica que la serie presenta una estructura autoregresiva y de medias m贸viles de primer orden.

Dado este comportamiento, se explorar谩n modelos ARIMA con diferentes configuraciones de par谩metros para evaluar cu谩l se ajusta mejor a la serie de tiempo:

* **ARIMA(1,1,1)**: Considera tanto un t茅rmino autoregresivo ($p=1$) como un t茅rmino de media m贸vil ($q=1$)
* **ARIMA(0,1,1)**: Modelo basado 煤nicamente en la componente de media m贸vil ($q=1$), 煤til si la serie sigue un comportamiento de ruido blanco tras la diferenciaci贸n.
* **ARIMA(1,1,0)**: Modelo basado 煤nicamente en la componente autoregresivo ($p=1$), 煤til si la serie tiene memoria de sus valores pasados sin la necesidad de un t茅rmino MA.

```python
# Aseg煤rate de que no haya valores NaN en la columna antes de graficar
serie = data_arima["diff_y"]

# Crear subplots para ACF y PACF
fig, axes = plt.subplots(1, 2, figsize = (16, 6))

# Gr谩fico de la ACF
plot_acf(serie, lags = 20, ax = axes[0])
axes[0].set_title("ACF de la Diferenciaci贸n para Dep贸sitos s/Int de Bancos Medianos")

# Gr谩fico de la PACF
plot_pacf(serie, lags = 20, ax = axes[1])
axes[1].set_title("PACF de la Diferenciaci贸n para Dep贸sitos s/Int de Bancos Medianos")

# Mostrar las gr谩ficas
plt.tight_layout()
plt.show()
```

#### Construcci贸n de modelos ARIMA y la significancia de los par谩metros $p$, $d$ y $q$
$$p-value=\begin{cases} \text{Significativo} & \text{if } p-value \geq 0.05 \\ \text{No Significativo} & \text{En otro caso} \end{cases}$$

### Modelo 1: ARIMA(1,1,1)




Se analiza la significancia de los par谩metros por medio de los p-values:

* **ar.L1**: p-value = 0.000 -> Significativo
* **ma.L1**: p-value = 0.000 -> Significativo

Por lo tanto, se acepta este modelo 

```python
# Modelo 1
modelo = ARIMA(data_arima["y"], order = (1, 1, 1))
mod_salida = modelo.fit()
print(mod_salida.summary())
```

### Modelo 2: ARIMA(1,1,0)

Se analiza la significancia de los par谩metros por medio de los p-values:

* **ar.L1**: p-value = 0.624  -> NO Significativo

Por lo tanto, se descarta este modelo 

```python
# Modelo 2
modelo = ARIMA(data_arima["y"], order = (1, 1, 0))
mod_salida = modelo.fit()
print(mod_salida.summary())
```

### Modelo 3: ARIMA(0,1,1)

Se analiza la significancia de los par谩metros por medio de los p-values:

* **ma.L1**: p-value = 0.697  -> NO Significativo

Por lo tanto, se descarta este modelo 

```python
# Modelo 3
modelo = ARIMA(data_arima["y"], order = (0, 1, 1))
mod_salida = modelo.fit()
print(mod_salida.summary())
```

### Selecci贸n del Modelo

Tras la evaluaci贸n de los modelos propuestos **ARIMA(1,1,1), ARIMA(0,1,1) y ARIMA(1,1,0)**, se analiz贸 la significancia estad铆stica de sus par谩metros mediante la inspecci贸n de los p-values de sus coeficientes. De acuerdo con los resultados obtenidos, el modelo **ARIMA(1,1,1)** presenta coeficientes estad铆sticamente significativos, ya que sus p-values son menores a 0.05, lo que indica que los t茅rminos autoregresivos ($p=1$) y de media m贸vil ($q=1$) contribuyen significativamente a la explicaci贸n de la serie de tiempo.

Dado que la significancia de los par谩metros es un criterio fundamental para garantizar la robustez del modelo, y que este ha demostrado ser adecuado tanto en t茅rminos de ajuste como de capacidad explicativa, se selecciona ARIMA(1,1,1) como el modelo final para la generaci贸n de pron贸sticos (Forecast).

```python
# Modelo Seleccionado
modelo = ARIMA(data_arima["y"], order = (1, 1, 1))
mod_salida = modelo.fit()
```

### Evaluaci贸n del Modelo

**An谩lisis de Residuales**

Al tener un p-value = 1 > 0.05, significa que **NO hay Ruido Blanco**

En el test de Normalidad, se obtiene un p-value =  2.09246039e-10] < 0.05, lo que significa que **S existe Normalidad en los Residuales**

```python
plt.rcParams.update({"figure.figsize": (9, 3), "figure.dpi": 120})

residuals = pd.DataFrame(mod_salida.resid)[1:]
fig, ax = plt.subplots(1, 2)
residuals.plot(title = "Residuales", ax = ax[0], color = "orange", lw = 2)
residuals.plot(kind = "kde", title = "Densidad", ax = ax[1], color = "orange", lw = 2)
plt.show()

print(sm.stats.acorr_ljungbox(mod_salida.resid, lags = [10], return_df = True))
print(normaltest(residuals))
```

### Efectividad del Modelo con Base en el $R^2$
 

El coeficiente de determinaci贸n $R^2$ obtenido para el modelo **ARIMA(1,1,1)** es del **96.34%**, lo que indica que el modelo es altamente efectivo para explicar la variabilidad de la serie de tiempo. Este valor sugiere que el modelo captura de manera precisa la estructura de los datos, minimizando el error y proporcionando estimaciones confiables.

Un $R^2$ elevado implica que el modelo ajusta correctamente la serie y que la relaci贸n entre los valores observados y los valores ajustados es consistente. Adem谩s, la comparaci贸n con modelos alternativos confirma que ARIMA(1,1,1) logra una mejor representaci贸n de los patrones subyacentes en los datos, lo que refuerza su validez para realizar pron贸sticos futuros.

Por lo tanto, debido a su alto nivel de ajuste y capacidad predictiva, se concluye que ARIMA(1,1,1) es un modelo adecuado para la generaci贸n de Forecasts, asegurando que las proyecciones sean representativas de la din谩mica de la serie analizada.

```python
# Predicciones dentro de la muestra (fitted values)
predicciones_in_sample = mod_salida.fittedvalues[1:]

# Excluir el primer valor por diferenciaci贸n
reales = data_arima["y"].iloc[1:]

# Calcular R虏 in-sample
r2_in_sample = r2_score(reales, predicciones_in_sample)
print(f"R虏 dentro de la muestra (in-sample): {r2_in_sample}")
```

## 3锔 Predicciones con XGBoost Regressor

```python
data_xgboost = data.copy()

# N煤mero de rezagos como variables predictoras
lags = 5
for i in range(1, lags + 1):
    data_xgboost[f"lag_{i}"] = data_xgboost["y"].shift(i)

data_xgboost = data_xgboost.dropna().reset_index(drop=True)

data_xgboost.head()
```

```python
data_xgboost = data.copy()

# N煤mero de rezagos como variables predictoras
lags = 5
for i in range(1, lags + 1):
    data_xgboost[f"lag_{i}"] = data_xgboost["y"].shift(i)

# Eliminar valores NaN generados por los lags
data_xgboost = data_xgboost.dropna().reset_index(drop=True)

# Divisi贸n de conjuntos Train y Test
train_size = int(len(data_xgboost) * 0.8)  # 80% para entrenamiento
train, test = data_xgboost.iloc[:train_size], data_xgboost.iloc[train_size:]

# Creaci贸n de conjuntos de Entrenamiento y Pruebas
X_train, y_train = train.drop(columns = ["FECHA", "y"]), train["y"]
X_test, y_test = test.drop(columns = ["FECHA", "y"]), test["y"]

# Entrenamiento de Modelo
modelo_xgb = XGBRegressor(n_estimators = 50, learning_rate = 0.2, objective = "reg:squarederror")
modelo_xgb.fit(X_train, y_train)

# Predicciones dentro de la Serie
pred_train = modelo_xgb.predict(X_train)
pred_test = modelo_xgb.predict(X_test)
pred_series = np.concatenate([pred_train, pred_test])

min_length = min(len(data_xgboost["y"].iloc[lags:]), len(pred_series))
y_real = data_xgboost["y"].iloc[lags:lags+min_length].values
pred_series = pred_series[:min_length]
r2_test = r2_score(y_test, pred_test)
r2_completo = r2_score(y_real, pred_series)

# Predicciones fuera de la Serie (Rolling Forecast)
n_forecast = 12

last_known_data = test.iloc[-1:].copy()

future_predictions = []

for i in range(n_forecast):
    X_input = last_known_data.drop(columns=["FECHA", "y"]).values  # Eliminar la columna de fecha para predecir
    next_pred = modelo_xgb.predict(X_input)[0]  # Hacer la predicci贸n
    future_predictions.append(next_pred)  # Guardar el valor predicho

    # Actualizar los lags con la nueva predicci贸n
    for j in range(lags, 0, -1):
        last_known_data[f"lag_{j}"] = last_known_data[f"lag_{j-1}"] if j > 1 else next_pred

future_dates = pd.date_range(start=test["FECHA"].iloc[-1], periods=n_forecast + 1, freq="M")[1:]
df_future = pd.DataFrame({"FECHA": future_dates, "Predicciones_Futuras": future_predictions})

# Gr谩ficos de la Serie de Tiempo
plt.figure(figsize=(12,6))

plt.plot(data_xgboost["FECHA"], data_xgboost["y"], label="Datos Originales", marker="o", color="black")

plt.plot(test["FECHA"], pred_test, label="Predicciones XGBoost (Test)", linestyle="dashed", color="red")

plt.plot(df_future["FECHA"], df_future["Predicciones_Futuras"], label="Forecast XGBoost (12 meses)", linestyle="dashed", color="blue")

plt.xlabel("Fecha")
plt.ylabel("Valores")
plt.title(f"Predicci贸n con XGBoost\nR虏 Test: {r2_test:.4f} | R虏 Completo: {r2_completo:.4f}")
plt.legend()
plt.show()

df_predicciones = data_xgboost.iloc[lags:lags+min_length].copy()
df_predicciones["Predicciones_XGBoost"] = pred_series

print(f"R虏 en Test: {r2_test:.4f}")
print(f"R虏 en toda la serie: {r2_completo:.4f}")
```

```python
plt.figure(figsize=(12,6))
plt.plot(data_xgboost["FECHA"], data_xgboost["y"], label="Valores Reales", marker="o", color="black")
plt.plot(df_predicciones["FECHA"], df_predicciones["Predicciones_XGBoost"], label="Predicciones Dentro de la Muestra", linestyle="dashed", color="red")
plt.plot(df_future["FECHA"], df_future["Predicciones_Futuras"], label="Forecast XGBoost (12 meses)", linestyle="dashed", color="blue")
plt.xlabel("Fecha")
plt.ylabel("Valores")
plt.title("Valores Reales + Predicciones Fuera de la Serie")
plt.legend()
plt.show()
```
